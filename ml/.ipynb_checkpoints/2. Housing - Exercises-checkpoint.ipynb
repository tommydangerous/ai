```python
DOWNLOAD_ROOT = 'https://raw.githubusercontent.com/ageron/handson-ml/master/'
HOUSING_PATH = 'datasets/housing'
HOUSING_URL = DOWNLOAD_ROOT + HOUSING_PATH + '/housing.tgz'

import numpy as np
import os
import pandas as pd
import tarfile
from six.moves import urllib

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, 'housing.csv')
    return pd.read_csv(csv_path)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
    
def detect(item, array):
    try:
        return array[array.index(item)]
    except ValueError:
        return None
```

```python
housing_master = load_housing_data()
original_attributes = list(housing_master)
```

```python
from pandas.plotting import scatter_matrix

%matplotlib inline
import matplotlib.pyplot as plt

attributes = [
    'median_house_value',
    'median_income',
    'total_rooms',
    'housing_median_age',
]
scatter_matrix(housing_master[attributes], figsize=(12, 8))
```

```python
# Looking for correlations
correlation_matrix = housing_master.corr()
correlation_matrix['median_house_value'].sort_values(ascending=False)

housing_master['rooms_per_household'] = (
    housing_master['total_rooms'] / housing_master['households']
)
housing_master['bedrooms_per_room'] = (
    housing_master['total_bedrooms'] / housing_master['total_rooms']
)
housing_master['population_per_household'] = (
    housing_master['population'] / housing_master['households']
)
correlation_matrix2 = housing_master.corr()
correlation_matrix2['median_house_value'].sort_values(ascending=False)
```

```python
# Stratifying

# Stratify using median income since experts told us ahead of time
# that median income is very important to predicting housing prices.
housing_master['income_category'] = np.ceil(
    housing_master['median_income'] / 1.5
)
housing_master['income_category'].where(
    housing_master['income_category'] < 5,
    5.0,
    inplace=True,
)

from sklearn.model_selection import StratifiedShuffleSplit

stratified_split = StratifiedShuffleSplit(
    n_splits=1,
    test_size=0.2,
    random_state=42,
)
# for train_index, test_index in 
gen = stratified_split.split(
    housing_master,
    housing_master['income_category'],
)
for training_indices, test_indices in gen:
    training_set = housing_master.loc[training_indices]
    test_set = housing_master.loc[test_indices]
    
series = [
    data['income_category'].value_counts() / len(data) for data in
    [housing_master, training_set, test_set]
]

# Even distribution
for i in range(1, 6):
    print(i, [ser.loc[i] for ser in series])
```

```python
# Data Cleaning
from sklearn.preprocessing import Imputer

training_set_cleaned = training_set.drop('median_house_value', axis=1)
training_set_labels = training_set['median_house_value'].copy()

test_set_cleaned = test_set.drop('median_house_value', axis=1)
test_set_labels = test_set['median_house_value'].copy()

imputer = Imputer(strategy='median')

housing_numerical_attrs_only = training_set_cleaned.drop(
    'ocean_proximity',
    axis=1,
)

imputer.fit(housing_numerical_attrs_only)
imputer.statistics_ == housing_numerical_attrs_only.median().values

no_missing_data = imputer.transform(housing_numerical_attrs_only)
no_missing_data_data_frame = pd.DataFrame(
    no_missing_data,
    columns=housing_numerical_attrs_only.columns,
)
```

```python
# Encoding
from sklearn.preprocessing import LabelBinarizer

ocean_proximity = training_set_cleaned['ocean_proximity']
encoder = LabelBinarizer(sparse_output=True)
housing_cat_1hot = encoder.fit_transform(ocean_proximity)
housing_cat_1hot[0]
```

```python
# Preparation
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import FeatureUnion, Pipeline
from sklearn.preprocessing import StandardScaler

# These attributes are appended to the prepared data via the 
# pipeline transformations
housing_numerical_attrs_extra = [
    'rooms_per_household',
    'population_per_household',
    'bedrooms_per_room',
]
numerical_attributes = list(
    filter(
        lambda f: not detect(f, housing_numerical_attrs_extra),
        list(housing_numerical_attrs_only),
    ),
)
all_numerical_attributes = numerical_attributes + housing_numerical_attrs_extra
categorical_attributes = ['ocean_proximity']

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, extra_features=housing_numerical_attrs_extra):
        self.extra_features = extra_features
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        rooms_index = 3
        bedrooms_index = 4
        population_index = 5
        household_index = 6
        
        valid_features = list(
            filter(
                lambda f: detect(f, housing_numerical_attrs_extra),
                self.extra_features,
            )
        )
        
        arr = np.c_[X]
        
        if len(valid_features) >= 1:
            if detect('rooms_per_household', valid_features):
                rooms_per_household = X[:, rooms_index] / X[:, household_index]
                arr = np.c_[arr, rooms_per_household]
                
            if detect('population_per_household', valid_features):
                population_per_household = (
                    X[:, population_index] / X[:, household_index]
                )
                arr = np.c_[arr, population_per_household]
                
            if detect('bedrooms_per_room', valid_features):
                bedrooms_per_room = (
                    X[:, bedrooms_index] / X[:, rooms_index]
                )
                arr = np.c_[arr, bedrooms_per_room]

        return arr

class DateFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        return X[self.attribute_names].values
    
def build_pipeline(numerical_attributes):
    numerical_pipeline = Pipeline([
        ('selector', DateFrameSelector(numerical_attributes)),
        ('imputer', Imputer(strategy='median')),
        ('attributes_adder', CombinedAttributesAdder()),
        ('standard_scalar', StandardScaler()),
    ])

    categorical_pipeline = Pipeline([
        ('selector', DateFrameSelector(categorical_attributes)),
        ('label_binarizer', LabelBinarizer(sparse_output=True)),
    ])

    full_pipeline = FeatureUnion(transformer_list=[
        ('numerical_pipeline', numerical_pipeline),
        ('categorical_pipeline', categorical_pipeline),
    ])
    
    return full_pipeline

encoder = LabelBinarizer()
encoder.fit_transform(training_set_cleaned[categorical_attributes])

category_one_hot_attributes = list(encoder.classes_)
all_attributes = all_numerical_attributes + category_one_hot_attributes

pipeline = build_pipeline(numerical_attributes)

# 17 features: 9 from training_set_cleaned, 3 from attributes adder, 5 from label binarizer
training_set_prepared = pipeline.fit_transform(training_set_cleaned)
test_set_prepared = pipeline.transform(test_set_cleaned)
```

`3.` Try adding a transformer in the preparation pipeline to select
only the most important attributes.

```python
feature_importances = sorted(
    zip(dt_model.feature_importances_, all_attributes),
    reverse=True,
)
feature_limit = 7
top_features = []
for t, f in feature_importances[:feature_limit]:
    top_features.append(f)
top_features
```

```python
# Select the categorical top features
class TopFeatureCategoryTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, features):
        self.features = features
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        top_categorical_features = list(
            filter(
                lambda f: detect(f, self.features),
                category_one_hot_attributes,
            ),
        )
        
        categories_transformed = []
        for cat in X:
            cop = cat.copy()
            if not detect(cop, top_categorical_features):
                cop[0] = ''
            categories_transformed.append(cop)

        return np.array((categories_transformed))
```

```python
all_numerical_top_features = list(
    filter(lambda f: detect(f, all_numerical_attributes), top_features)
)
extra_top_features = list(
    filter(lambda f: detect(f, housing_numerical_attrs_extra), top_features)
)
numerical_pipeline = Pipeline([
    ('selector', DateFrameSelector(all_numerical_top_features)),
    ('imputer', Imputer(strategy='median')),
    ('standard_scalar', StandardScaler()),
])

categorical_pipeline = Pipeline([
    ('selector', DateFrameSelector(categorical_attributes)),
    (
        'top_feature_category',
        TopFeatureCategoryTransformer(top_features),
    ),
    ('label_binarizer', LabelBinarizer()),
])

transformer_list = []

if len([i for i in top_features if i in all_numerical_top_features]) >= 1:
    transformer_list.append(('numerical_pipeline', numerical_pipeline))

if len([i for i in top_features if i in category_one_hot_attributes]) >= 1:
    transformer_list.append(('categorical_pipeline', categorical_pipeline))

pipeline_tf = FeatureUnion(transformer_list=transformer_list)
training_set_prepared_tf = pipeline_tf.fit_transform(training_set_cleaned)
test_set_prepared_tf = pipeline_tf.transform(test_set_cleaned)
```

```python
from scipy.stats import randint as sp_randint

dt_tf = DecisionTreeRegressor()
dt_tf_param_grid = {
    'max_features': sp_randint(1, feature_limit),
}

dt_tf_search = RandomizedSearchCV(
    dt_tf,
    dt_tf_param_grid,
    cv=5,
    n_iter=7,
    random_state=42,
    scoring='neg_mean_squared_error',
)

dt_tf_search.fit(training_set_prepared_tf, training_set_labels)
```

```python
dt_tf_model = dt_tf_search.best_estimator_
dt_tf_predictions = dt_tf_model.predict(test_set_prepared_tf)
dt_tf_mse = mean_squared_error(test_set_labels, dt_tf_predictions)
# 6 - 70651.38169018233
# 7 - 66961.791616011251
# 8 - 70410.221040221833
# 9 - 68352.236212098607
# 10 - 71105.196908644808
# 11 - 70339.710915549978
# 12 - 70839.227096546892
# 13 - 70942.851367650001
#=> When using 7 of top features, we get the best error of 66961.79
dt_tf_rmse = np.sqrt(dt_tf_mse)
dt_tf_rmse
```
